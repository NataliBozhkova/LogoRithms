```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.countplot(y="What company name do you think was used for the creation of the following logos?", data=data)
plt.title("Question 1")
plt.ylabel("Question")
plt.savefig("question")
```

```python
# We created figures for each question to see the results of all of them if needed
for i in range(40):
  if i == 0:
    col = "What company name do you think was used for the creation of the following logos?"
  else:
    col = f"What company name do you think was used for the creation of the following logos?.{i}"
  sns.countplot(y=col, data=data)
  plt.title(f"Question {i+1}")
  plt.ylabel("Question")
  plt.savefig(f"question_{i+1}.png", dpi=300)
```

```python
import re
scores = []

for score_string in data["Total score"].values:   #We cleaned the data keeping the scores
  score = int(score_string.split(".")[0])
  scores.append(score)
```

```python
# We plotted the distribution of the points
import numpy as np

ax = sns.histplot(x=scores, bins=list(range(41)), color = "purple")
plt.title("Distribution of scores")
plt.xlabel("Score")

string = f"Mean: {np.round(np.mean(scores), 2)}\nMedian: {np.round(np.median(scores), 0)}"

plt.text(x=1, y=1.1, s=string, transform=ax.transAxes, horizontalalignment='right', verticalalignment='center', color="blue")
```
